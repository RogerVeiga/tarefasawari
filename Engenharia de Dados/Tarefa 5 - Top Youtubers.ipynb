### 1 - Tratar as colunas sem informação adicionando o texto “Não informado”.

```python
import pandas as pd
df = pd.read_csv('arquivos/topyoutube.csv')
df.head()
```

```python
df = df.fillna('Não informado')
df.loc[1582]
```

### 2 - Formatar a coluna de total de inscritos multiplicando por 100 milhões.

```python
df
```

```python
# Removendo as vírgulas e convertendo a coluna 'Total Views' para float
df['Total Views'] = df['Total Views'].str.replace(',', '').astype(float) * 100000000
```

```python
df.head()
```

### 3 - Formatar coluna “AVG” para 2 casas decimais depois da vírgula.

```python
df['Avg'] = df['Avg'].round(2)
df.head()
```
